```python
import ipymd
```

```python
data = ipymd.data_input.cif.CIF()
data.setup_data('Cu.cif')
```

```python
atoms = data.get_atom_data()
print(atoms)
```

```python
meta = data.get_meta_data()
print(meta)
```

```python
vis = ipymd.visualise_sim.Visualise_Sim()
vis.basic_vis(data.get_atom_data(), data.get_meta_data(), xrot=45,yrot=45)
```

### Tests directly from the ipymd website

```python
cif_path = ipymd.get_data_path('example_crystal.cif')
data = ipymd.data_input.cif.CIF()
data.setup_data(cif_path)
meta = data.get_meta_data()
vis = ipymd.visualise_sim.Visualise_Sim()
vis.basic_vis(data.get_atom_data(), data.get_meta_data(), xrot=45,yrot=45)
```

```python
import ipymd
import pandas as pd
df = pd.DataFrame(\
    [[2,3,4,1,[0, 0, 255],1],\
    [1,3,3,1,'orange',1],\
    [4,3,1,1,'blue',1]],\
    columns=['x','y','z','radius','color','transparency'])
vis = ipymd.visualise_sim.Visualise_Sim()
vis.add_atoms(df)
img1 = vis.get_image(size=400,quality=5)
img1
```
